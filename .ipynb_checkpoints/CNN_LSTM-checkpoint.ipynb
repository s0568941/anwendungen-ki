## Implementation of The Dataset
To be able to work with large amounts of data, the pytorch utils library provides the `Dataset` class. Derivatives of this class allow efficient use of large data sets by splitting data into batches which are sequentially supplied to the network.
The implemented dataset generates batches and deals with normalization.

```python
import torch
from torch.utils.data import Dataset
import os
import random
import pickle
import numpy as np
from torch.utils.data import DataLoader


# input parameters:
# features: array of strings, containing the needed features, e.g. ['Close', 'Volume']
# normalize_data: boolean, toggles normalization of data
class StockDatasetTrain(Dataset):
    def __init__(self, features, normalize_data=False):
        # array of strings containing the features which need to be extracted
        self.features = features
        self.normalize_data = normalize_data
        # the given path contains as a list all training samples
        # each entry will be accessed via `idx` in __getitem__
        if normalize_data:
            self.paths = self.listdir_nohidden(path_to_normalized_data_train)
        else:   
            self.paths = self.listdir_nohidden(path_to_processed_data_train)
        random.shuffle(self.paths)
        
    # avoid loading hidden folders which start with a dot, e.g. .DS_Store on Mac OS
    def listdir_nohidden(self, path):
        files = list()
        for file in os.listdir(path):
            if not file.startswith('.'):
                files.append(file)
        return files
        
    def __len__(self):
        return len(self.paths)
    
    def __getitem__(self, idx):
        with open(f'{path_to_processed_data_train}/{self.paths[idx]}', 'rb') as f:
            item = pickle.load(f)
            
        # retrieve the independent variables
        # array self.features determines which columns need to be returned
        data = item[self.features]
        # retrieve the dependent variables
        # target variable 'class' is the label
        label = item['class'].values
        label = label[len(label)-1]

        x = torch.tensor(data[self.features].values)
        
        # when using CrossEntropyLoss: one hot encoding must not define the target 
        # instead; index is used
        # label for sequence is the calculated class on the last day of the sequence
        # y will be the index for that particular class from the list CLASSES_TO_PREDICT
        y = torch.tensor([CLASSES_TO_PREDICT.index(item['class'].values[SEQUENCE - 1])])
        return x,y
    
# input parameters:
# features: array of strings, containing the needed features
# normalize_data: boolean, toggles normalization of data
class StockDatasetValidation(Dataset):
    def __init__(self, features, normalize_data=False):
        # array of strings containing the features which need to be extracted
        self.features = features
        self.normalize_data = normalize_data
        # the given path contains as a list all validation samples
        # each entry will be accessed via `idx` in __getitem__
        if self.normalize_data:
            self.paths = self.listdir_nohidden(path_to_normalized_data_val)
        else:    
            self.paths = self.listdir_nohidden(path_to_processed_data_val)
        random.shuffle(self.paths)
        
    # avoid loading hidden folders which start with a dot, e.g. .DS_Store on Mac OS
    def listdir_nohidden(self, path):
        files = list()
        for file in os.listdir(path):
            if not file.startswith('.'):
                files.append(file)
        return files
        
    def __len__(self):
        return len(self.paths)
    
    def __getitem__(self, idx):
        with open(f'{path_to_processed_data_val}/{self.paths[idx]}', 'rb') as f:
            item = pickle.load(f)
            
        # retrieve the independent variables
        # array self.features determines which columns need to be returned
        data = item[self.features]
        # retrieve the dependent variables
        # target variable 'class' is the label
        label = item['class'].values
        label = label[len(label)-1]
        
        x = torch.tensor(data[self.features].values)
        
        # when using CrossEntropyLoss: one hot encoding must not define the target 
        # instead; index is used
        # label for sequence is the calculated class on the last day of the sequence
        # y will be the index for that particular class from the list CLASSES_TO_PREDICT
        y = torch.tensor([CLASSES_TO_PREDICT.index(item['class'].values[SEQUENCE - 1])])
        return x,y
 
```

## Preprocessing of Raw Data

This cell will access the raw csv stock data files.
Working with `pandas`, two columns will be generated.
Column `avg_future_to_predict` contains the average closing price for the next `future_period_predict` days (as described in the research paper).
Column `class` will contain the target variable which will be predicted. This is derived by the `CLASSES_TO_PREDICT` which were defined in the research paper as well.
This way, each row of the data (which defines one trading day) will contain a class which represents the average earning rate for the next three days after that particular day.

### Set a Number for How Many Files to Preprocess
If, for example for testing this notebook, you do not want to use all 12,000+ csv files;
Set `limit_of_files_to_process` to the number of files you wish to preprocess.

```python
import torch
import os
import random
import pickle
import numpy as np
import pandas as pd
import time

# determines for each day the earning rate class label
# params:
# last_day_of_seq (float): closing price of last day of specific sequence - for preprocessing it will use all closing prices
# avg_future_to_predict (float): avg future closing price for last_day_of_seq
# returns label defining the earning rate class for given pair
def classify(last_day_of_seq, avg_future_to_predict):

    earning_rate = ((float(avg_future_to_predict) / float(last_day_of_seq)) - 1) * 100

    for label_class in CLASSES_TO_PREDICT:
        if isinstance(label_class, tuple):
            begin, end = label_class # unpack the tuple
            if begin <= earning_rate <= end:
                return label_class
        else: # if it is not a tuple, it is the class on the highest / lowest range
            if label_class < 0 and earning_rate <= label_class:
                return label_class
            elif label_class > 0 and earning_rate >= label_class:
                return label_class


# processes csv files into pickle files - each pickle contains one sequence of SEQUENCE days
# params: 
# normalize; flag which toggles normalization of data
# limit_of_files_to_process (int); set a limit of how many files you wish to process
def process_data(normalize=False, limit_of_files_to_process=None):
    files = os.listdir(path_to_raw_data)
    files = sorted(files, key=str.lower)

    counter = 1 # keep track how many files have been processed
    # do not include hidden files which start with a dot
    for file in files:
        if file.startswith('.'):
            continue
        # sets a limit to how many files will be processed by the given parameter
        limit_of_files_to_process = len(files) + 5 if limit_of_files_to_process is None else limit_of_files_to_process
        if counter > limit_of_files_to_process:
            return
        df = pd.read_csv(f'{path_to_raw_data}/{file}')
        counter +=1
        
        # makes sure that entries are ordered by the date - but is not run, as discussed in review meeting on May 7th
        #df.sort_values(by='Date')
                
        # Column with future (close) prices
        # each row will now have the closing price from `future_period_predict` days in the future
        df['avg_future_to_predict'] = df['Close'].shift(-future_period_predict)
        # last future_period_predict rows will contain NaN (due to shift) which need to be removed
        df.dropna(inplace=True)
        # for each time point, the average closing price of next `future_period_predict` is calculated
        df['avg_future_to_predict'] = df['avg_future_to_predict'].rolling(future_period_predict).mean()
        # first (future_period_predict - 1) rows will contain NaN (due to calculation where each row needs the previous values)
        df.dropna(inplace=True)        
        
        # introduce column which classifies the earning rate (Paper suggested 7 categories)
        df['class'] = list(map(classify, df['Close'], df['avg_future_to_predict']))
        
        # this methodology is used by the paper and called 'normalization', even though this method is more known as standardization
        if normalize:
            # Standardization - mean = 0 and std = 1 for the given dataset
            mean, std = np.mean(df[['Open', 'High', 'Low', 'Close', 'Volume']]), np.std(df[['Open', 'High', 'Low', 'Close', 'Volume']])
            df[['Open', 'High', 'Low', 'Close', 'Volume']] = df[['Open', 'High', 'Low', 'Close', 'Volume']] - mean
            df[['Open', 'High', 'Low', 'Close', 'Volume']] = df[['Open', 'High', 'Low', 'Close', 'Volume']] / std
            
            # Normalization - scaling values between 0 and 1
            #df[['Open', 'High', 'Low', 'Close', 'Volume']] = (
            #    df[['Open', 'High', 'Low', 'Close', 'Volume']] - df[['Open', 'High', 'Low', 'Close', 'Volume']].min()) / (
            #    df[['Open', 'High', 'Low', 'Close', 'Volume']].max() - df[['Open', 'High', 'Low', 'Close', 'Volume']].min()
            #)

            
        
        # dictionary is used for pickling the sequential data
        data = dict()
        # split data into training and validation sets
        # Paper suggested a ratio of 25% (validation) and 75% (training)
        end_of_train_data = int(train_split * len(df.index.values))
        # first part of data is used for training (first 75%)
        data['train'] = df[0:end_of_train_data]
        # second part of data is used for validation (last 25%)
        data['validation'] = df[end_of_train_data:]
        
        # removes path and determines filename without extension
        file_name = file.split('/')
        file_name = file_name[len(file_name) - 1].split('.')[0]
        
        # setup directories for processed data
        # normalized data is separated from unnormalized data
        train_dir = f'{path_to_processed_data}/normalized/{SEQUENCE}_day_sequences/train' if normalize else f'{path_to_processed_data}/{SEQUENCE}_day_sequences/train'
        val_dir = f'{path_to_processed_data}/normalized/{SEQUENCE}_day_sequences/validation' if normalize else f'{path_to_processed_data}/{SEQUENCE}_day_sequences/validation'
        dirs = [train_dir, val_dir]   
        # if directories do not exist; they are created now
        for target_dir in dirs:
            if not os.path.exists(target_dir):
                os.makedirs(target_dir)
        
        
        # generates pickles for training
        # to make sure that all samples have same size; excess data from the past is removed
        # i.e.: a file may have 124 rows (days) 
        # - dividing it into batches of 30 sequences means that 4 rows remain - these are therefore removed
        excess_train_data = len(data['train'].index.values) % SEQUENCE
        data['train'] = data['train'][excess_train_data:]

        # generate one pickle file containing one sequence of 30 days
        for seq in range(SEQUENCE, (len(data['train'].index.values) + 1), SEQUENCE):
            start_date_index = seq - SEQUENCE 

            # save training data for each sequence with name containing start date of sequence
            train_seq_start_date = str(data['train'].iloc[[start_date_index]]['Date'].values[0])
            with open(f'{train_dir}/{file_name}_{train_seq_start_date}.pkl', 'wb') as f:
                pickle.dump(data['train'][start_date_index:seq], f, pickle.HIGHEST_PROTOCOL)
            
            
        # generates pickles for validation
        # to make sure that all samples have same size; excess data from the past is removed
        excess_valid_data = len(data['validation'].index.values) % SEQUENCE
        data['validation'] = data['validation'][:-excess_valid_data]

        for seq in range(SEQUENCE, (len(data['validation'].index.values) + 1), SEQUENCE):

            start_date_index = seq - SEQUENCE 
            # save training data for each sequence
            val_seq_start_date = str(data['validation'].iloc[[start_date_index]]['Date'].values[0])
            with open(f'{val_dir}/{file_name}_{val_seq_start_date}.pkl', 'wb') as f:
                pickle.dump(data['validation'][start_date_index:seq], f, pickle.HIGHEST_PROTOCOL)
                      
        
print("Processing ...")       
#process_data(limit_of_files_to_process=None) # processed data is not normalized
#process_data(normalize=True, limit_of_files_to_process=None) # processed data is normalized
print(f'Saved to {path_to_processed_data}/{SEQUENCE}_day_sequences/')
print("End")
```

## Implementation of The Neural Network From The Research Paper
The researchers suggested as an effective neural network an LSTM model which consists of:
- 3 LSTM layers
- 1 linear output layer

The aim of this model is the prediction of earning rate class (out of 7 classes) per sequence (sequence of 30 days)

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

# params:
# input_size (int): number of features
# hidden_size (int): number of features in the hidden state h
# num_layers (int): number of recurrent LSTM layers 
# num_directions (int): unidirectional or bidirectional
# batch_size (int): batch size
class Net(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_directions, batch_size):
        super(Net, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.num_directions = num_directions
        self.batch_size = batch_size
        self.num_classes = len(CLASSES_TO_PREDICT)
        # layers:
        self.stacked_lstm = nn.LSTM(self.input_size, self.hidden_size, self.num_layers, batch_first=True)
        self.linear = nn.Linear(self.hidden_size, self.num_classes)
        
    def forward(self, x):
        # hidden state: initialized as empty state / zeroes
        # h_n of shape (num_layers * num_directions, batch, hidden_size)
        h0 = torch.zeros(self.num_layers * self.num_directions, self.batch_size, self.hidden_size) # initialize hidden state
        # c_n of shape (num_layers * num_directions, batch, hidden_size):
        c0 = torch.zeros(self.num_layers * self.num_directions, self.batch_size, self.hidden_size) # initialize cell state
        
        # forward pass of data through LSTM layers
        output, (hn, cn) = self.stacked_lstm(x, (h0, c0))
        
        # linear layer receives computation of last step from LSTM layer
        output = self.linear(output[:, -1, :])
        # nn.CrossEntropyLoss already applies softmax activation
        # therefore softmax activation must not be applied in forward method
        return output
```

## This architecture extends the LSTM model given by the paper.
It is extended by a convolutional neural network, consisting of three convolutional 1D layers, each one of them is followed by a max pool 1d and an activation function of type rectified linear.

The first version of this network would use `torch.relu` (regular relu function) for all layers. This however lead to a decrease in performance, which was the result of the vanishing gradient problem, caused by combining convolutions with relu activation functions. Hence, only the first layer applies `torch.relu` while the subsequent layers apply `torch.rrelu` ('leaky relu') to avoid the vanishing gradient problem.
This exact same problem was discovered by researches in the following research paper: https://iopscience.iop.org/article/10.1088/1757-899X/435/1/012026/pdf

```python
import torch
import torch.nn as nn

# input_size is the number of features
class CNN_LSTM_Net(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_directions, batch_size, incr_num_features = False):
        super(CNN_LSTM_Net, self).__init__()        
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.num_directions = num_directions
        self.batch_size = batch_size
        self.num_classes = len(CLASSES_TO_PREDICT)
        
        # CNN layers to convolve input data
        # input size of LSTM part of network is the number of output channels of the CNN layers
        # input size of the CNN part of the network is the initial sequence size
        #self.cnn_output_size = 128 # with this, the number of sequences is increased from 30 to 128
        self.conv1 = nn.Conv1d(SEQUENCE, 32, 1) # this layer adds 2 sequences (from 30 to 32)
        self.conv2 = nn.Conv1d(32, 64, 1) # this layer adds 32 sequences (from 32 to 64)
        self.conv3 = nn.Conv1d(64, 128, 1) # this layer adds 64 sequences (from 64 to 128)
        self.conv3_1 = nn.Conv1d(64, 128, 1, padding=1) # this layer adds 64 sequences (from 64 to 128) and adds 2 features
        self.pool1 = nn.MaxPool1d(1, 1) # number of features remains the same

        # specifies whether or not the network should extract more features from the given input features
        self.incr_num_features = incr_num_features
        self.retain_num_features = not self.incr_num_features 
        
        if self.incr_num_features:
            self.input_size += 2 # input size needs to be adjusted, to correctly initialize the LSTM network
        
        # LSTM layers with linear output layer (this part is the exact architecture used in the paper - except for hidden size):
        self.stacked_lstm = nn.LSTM(self.input_size, self.hidden_size, self.num_layers, batch_first=True)
        self.linear = nn.Linear(self.hidden_size, self.num_classes)
        
    # this method is the complete forward pass through the CNN network and returns sequential data for the LSTM network
    def convolve_input(self, x):
        output = torch.relu(self.pool1(self.conv1(x))) # shape = (batch_size, 32, 5)
        # after first layer; leaky relu activation function is used to mitigate the vanishing gradient problem
        output = torch.rrelu(self.pool1(self.conv2(output))) # shape = (batch_size, 64, 5)
        if self.retain_num_features:
        #print(out)
            return torch.rrelu(self.pool1(self.conv3(output))) # shape = (batch_size, 128, 5)
        elif self.incr_num_features:
            return torch.rrelu(self.pool1(self.conv3_1(output))) # shape = (batch_size, 128, 7)
        
    def forward(self, x):
        
        x = self.convolve_input(x) # sequence size is increased to 128
        
        # hidden state: initialized as empty state / zeroes
        # h_n of shape (num_layers * num_directions, batch, hidden_size)
        h0 = torch.zeros(self.num_layers * self.num_directions, self.batch_size, self.hidden_size) # initialize hidden state
        # c_n of shape (num_layers * num_directions, batch, hidden_size):
        c0 = torch.zeros(self.num_layers * self.num_directions, self.batch_size, self.hidden_size) # initialize cell state
        # input x to the LSTM network is the convolved output of the CNN network
        output, (hn, cn) = self.stacked_lstm(x, (h0, c0))
        # linear layer receives computation of last step from LSTM layer
        output = self.linear(output[:, -1, :])
        # nn.CrossEntropyLoss already applies softmax activation
        # therefore, softmax activation must not be done in forward method
        return output
```

## Autoencoder
This network is trained on the training data only, and aims to minimise the loss between the input and the output.
The goal is a complete reconstruction of the input data, after passing the input data through the networks bottleneck where the data is compressed. After this bottleneck (layer conv2), reconstruction / decoding expands the data and transforms it back to its initial shape.
By then comparing the output with the input, the network learns how to compress / decode given input and provide a reduction of dimensionality while also keeping the most important information from the data.
To later be able to encode the trained data, the network is split into an encoder (see method `encode`) and a decoder (see method `decode`). 
After having trained the Autoencoder; calling the `encode` method on the input returns a compressed version of the input with reduction of dimensionality. This output can then be fed through the actual network which does the desired predictions.
The advantage of encoding data before passing it through the actual network is that by reducing the dimensionality, the neural network no longer needs to put as high of an emphasis on extracting the relevant information from great amounts of data as it is already available in a condensed form. Therefore, predictions can be more accurate and achieved in less time and with less data, as compared to regular methods where dimensionality reduction is not applied.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class Autoencoder_Net(nn.Module):
    def __init__(self):
        super(Autoencoder_Net, self).__init__()
        # encoder
        self.conv1 = nn.Conv1d(30, 24, 3, stride=1) # output shape: (batch_size, 24, 3)
        self.conv2 = nn.Conv1d(24, 16, 3, stride=1) # output shape: (batch_size, 16, 1)
        #self.conv3 = nn.Conv1d(16, 8, 1, stride=1) # output shape would have been: (batch_size, 8, 1) 
        # but 2 layers had greater accuracy
        
        # decoder
        #self.conv4 = nn.ConvTranspose1d(8, 16, 1, stride=1)
        self.conv5 = nn.ConvTranspose1d(16, 24, 3, stride=1) # output shape: (batch_size, 24, 3)
        self.conv6 = nn.ConvTranspose1d(24, 30, 3, stride=1) # output shape: (batch_size, 30, 5) # same shape as input

    # for batch size 64, sequence length 30, feature size 5:
    # reduced (encoded) down to [64, 16, 1]
    def encode(self, x):
        out = torch.relu(self.conv1(x)) # reduced down to 3 features and 24 sequences
        out = torch.relu(self.conv2(out)) # reduced down to 1 feature and 16 sequences
        return out
    
    # for encoded batch size 64, sequence length 16, feature size 1 as input to decode method:
    # decompressed (decoded) up to [64, 30, 5]
    def decode(self, x):
        out = torch.relu(self.conv5(x)) # increased up to 3 features and 24 sequences
        out = torch.relu(self.conv6(out)) # increased up to 5 features and 30 sequences
        return out

    def forward(self, x):
        # encode the data
        out = self.encode(x)
        # decode the data
        out = self.decode(out)
        # tanh as a final activation function resulted in the network not learning
        #out = torch.tanh(out) 
        # sigmoid as final activation function lead to best and most accurate results
        out = torch.sigmoid(out)
        # having a linear layer as final output layer also lead to the network not learning
        #out = torch.relu(self.linear(out)) 
        return out
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

# hyperparameter tuning
import optuna
from optuna.trial import TrialState

### AE_LSTM ###
# this loads in a pretrained autoencoder
# the method encode of this network can then be used to encode data for a different network
checkpoint_ae = torch.load('./checkpoint_AE_sigmoid_layers2_epoch200_acc_32.22_1621899788.pth')
ae_net = Autoencoder_Net()
ae_net.load_state_dict(checkpoint_ae['model_state'])
### AE_LSTM ###

# this method handles the forward pass to the network
# can be used to train the network / autoencoder or to validate a given network
# params:
# inputs: input (X) data to pass through network
# labels: labels (Y) data used to determine the loss and train the network
# train (boolean): flag which toggles training / validation mode for network
# train_autoencoder (boolean): flag which toggles training of autoencoder 
    # (treated differently as output will be compared to the input; labels would then be the input as well)
def pass_data(inputs, labels, train=False, train_autoencoder=False):
    if train:
        net.train()
        optimizer.zero_grad()
        

    
    #outputs = net(inputs.float())
    #print('And detail: ', torch.max(inputs[0].int(), dim=1))
    loss = None
    predicted = None
    running_correct = None
    if train_autoencoder:
        ### AE_LSTM
        encoded_input = ae_net.encode(inputs.float())
        outputs = net(encoded_input)
        ### AE_LSTM ###
        # calculate accuracy and loss for autoencoder
        loss = criterion(outputs.float(), inputs.float())
        # calculate how many correct predictions for current iteration of this epoch
        running_correct = (inputs.view_as(outputs) == outputs).sum().item()
    else :
        # keep track of prediction accuracy
        _, predicted = torch.max(outputs, 1) 
        loss = criterion(torch.squeeze(outputs), torch.squeeze(labels))
        # calculate how many correct predictions for current iteration of this epoch
        running_correct = (labels.view_as(predicted) == predicted).sum().item()

    # when validating; backpropagation is not done
    if train:
        loss.backward()
        optimizer.step()
        
    return running_correct, loss.item()

# this method is the training method
# loads training and validation data which are used over specified number of epochs to train the initialised network
# network, optimizer, loss function, data loader (among others) are all initialised outside this method
# params: 
# checkpoint_dir (string): path to checkpoint file which contains the saved model that will be loaded in
    # if None: network initialised outside this method will then be used
# trial: implicitly passed by the Optuna module when creating a study
# train_autoencoder (boolean): flag which toggles training of the initialised autoencoder network
def start_training(checkpoint_dir=None, trial=None, train_autoencoder=False):
    global num_epochs, model_name
    loss_vals = list()
    # when checkpoint_dir is given; checkpoint will contain saved network
    checkpoint = None
    step = 0 # tensorboard uses this to keep track of current iteration
    start_epoch = 0 # used when continuing training of a saved model
    # if checkpoint directory is given, saved data will be loaded in and updated
    if checkpoint_dir:
        checkpoint = load_model(checkpoint_dir)
        loss_vals = checkpoint['loss_vals']
        cp_epoch = checkpoint['epoch']
        start_epoch = cp_epoch # start epoch will be the epoch at which training of saved model ended
        num_epochs += cp_epoch # start from the saved epoch and increase by that amount
        model_name = checkpoint['model_name']
        step = checkpoint['step']
    
    net.train()
    # model.log will log model name, timestamps, epochs, steps, accuracies and losses (both training and validation) of the network
    with open('model.log', 'a') as f:
        for epoch in range(start_epoch, num_epochs):
            epoch_start = time.time()
            print(f'Epoch {epoch + 1} / {num_epochs}')           
            n_samples = 0
            running_loss_train = 0
            running_correct_train = 0

            loader_idx = 0
            for inputs, labels in loader:
                # train_correct: number of correct predicitions in training
                train_correct = None
                train_loss = None
                if train_autoencoder:
                    # loss for autoencoder is calculated by comparing loss of output with the actual input
                    # output has to be equal to inputs (or very close as information gets lost during encoding)
                    train_correct, train_loss = pass_data(inputs, inputs, train=True, train_autoencoder=train_autoencoder)
                else:
                    train_correct, train_loss = pass_data(inputs, labels, train=True, train_autoencoder=train_autoencoder)
                running_loss_train += train_loss
                ######################################################################
                # keep track of prediction accuracy
                running_correct_train += train_correct
                n_samples += labels.size(0)
                accuracy_train = (running_correct_train / n_samples) * 100

                # Tensorboard writer tracks for each step accuracies and losses
                writer.add_scalar('Loss/train', running_loss_train / n_samples, global_step=step)
                writer.add_scalar('Accuracy/train', running_correct_train / n_samples, global_step=step)
                step += 1

                # logging in model.log
                # log string contains: model_name,epoch,step,timestamp,accuracy_train,loss_train,validation_accuracy,validation_loss
                # log string is comma separated
                # collecting data for plotting; 10 times per epoch:
                if not train_autoencoder:
                    if loader_idx % int(len(loader) / 10) == 0:
                        validation_accuracy, validation_loss = start_validation()
                        log_str = f'{model_name},{epoch},{step},{round(time.time(),3)},'
                        log_str += f'{round(float(accuracy_train),2)},{round(float(running_loss_train / n_samples),4)},'
                        log_str += f'{round(float(validation_accuracy),2)},{round(float(validation_loss),4)}'
                        f.write(f'{log_str}\n')
                loader_idx += 1


            epoch_loss = running_loss_train / len(loader)
            print(f'Epoch loss: {str(epoch_loss)}')
            loss_vals.append(epoch_loss)
            print(f'Accuracy of the network: {round(accuracy_train, 3)} %')
            # Tensorboard: additional parameters are added for extra visualization features
            writer.add_hparams({
                'lr': learning_rate,
                'bsize': batch_size
            },
            {
                'hparam/accuracy': accuracy_train,
                'hparam/loss': epoch_loss
            })

            ####
            epoch_end = time.time()
            print(f"{round(float(epoch_end - epoch_start),2)} seconds for epoch {epoch + 1}")
            ####
            # Handle pruning based on the intermediate value.
            #### Optuna
            if trial:
                trial.report(accuracy_train, epoch)
                if trial.should_prune():
                    raise optuna.exceptions.TrialPruned()
            #### Optuna
        
    # finally, the model is saved together with other parameters such as epoch, step, accuracy and loss
    save_model(num_epochs, step, accuracy_train, loss_vals)
    return accuracy_train, loss_vals

# this method is the validation part for the network
# it is structured in a very similar way to the start_training() method
# the main difference here is that backpropagation is disabled which means that the network will not work when data is passed through
# returns validation accuracy and loss
def start_validation():
    loss_vals = list()
    # activate evaluation mode
    net.eval()
    
    n_samples = 0

    valid_running_loss = 0
    valid_running_correct = 0
    step = 0 # for tensorboard
    
    # ensures that backpropagation is disabled
    with torch.no_grad():   
        
        for inputs, labels in valid_loader:
            # train_correct: number of correct predicitions in validation
            valid_correct, valid_loss = pass_data(inputs, labels)
            valid_running_loss += valid_loss
            ######################################################################
            # keep track of prediction accuracy
            valid_running_correct += valid_correct
            n_samples += labels.size(0)

            # Tensorboard writer tracks for each step accuracies and losses
            writer.add_scalar('Loss/valid', valid_running_loss / n_samples, global_step=step)
            #writer.add_scalar('Accuracy/train', running_correct / len(loader), global_step=step)
            writer.add_scalar('Accuracy/valid', valid_running_correct / n_samples, global_step=step)
            step += 1                    
            
        validation_loss = valid_running_loss / len(loader)
        print(f'Validation loss: {str(validation_loss)}')
        loss_vals.append(validation_loss)
        accuracy = (valid_running_correct / n_samples) * 100
        print(f'Validation accuracy of the network: {round(accuracy, 3)} %')
        
        loss_vals.append(validation_loss)

    return accuracy, validation_loss
```

## Define Methods to Save And Load Model
When training a network takes to long, these methods are able to serialise and deserialise a trained model.
Thereby, a model can be saved after a number of epochs and then continue from there at a different time.

```python
import time

# saves model and other parameters related to training as a dictionary
# file name of checkpoint needs to be adjusted manually for each run - but can stay the same as no checkpoint will be overwritten
    # by including timestamps into file name
# params:
# epoch (int): epoch where training ended
# step (int): step where training ended
# accuracy (float): accuracy where training ended
# loss_vals (array of floats): loss values where training ended
def save_model(epoch, step, accuracy, loss_vals):

    checkpoint = {
        "epoch": epoch,
        "model_state": net.state_dict(),
        "optimizer_state": optimizer.state_dict(),
        "loss_vals": loss_vals,
        "model_name": model_name,
        "step": step
    }

    # file name includes epoch and accuracy (as an integer)
    checkpoint_path = f'checkpoint_AE_sig_l2_LSTM_NORM_REAL_epoch{epoch}_acc_{round(accuracy,2)}_{int(time.time())}.pth'
    torch.save(checkpoint, checkpoint_path)


# loads in the saved / serialised model (and optimizer) and returns the checkpoint as a dictionary
# params:
# checkpoint_path (string): path to serialised model
def load_model(checkpoint_path):
    global net, optimizer
    checkpoint = torch.load(checkpoint_path)
    net.load_state_dict(checkpoint['model_state'])
    optimizer.load_state_dict(checkpoint['optimizer_state'])

    return checkpoint
```

## Training for Method E3.1 - Autoencoder as Preprocessor to LSTM Network - Not Normalized
In this cell, the LSTM neural network defined in the paper is extended by a convolutional autoencoder.
This is the third extension `E3.1` of the research papers work.
The features which are passed to the network are:
`Open`, `High`, `Low`, `Close` and `Volume`
Data is not normalized.

As the previous extensions worsened the performance of the network by extracting more data through convolutions, this extension reuses the LSTM network of the research paper and extends it by applying a trained autoencoder. The autoencoder is used in a way where the input data is encoded by the autoencoder, reducing its dimensionality and then passing on the denoised data to the actual LSTM network.
The LSTM network will then receive data of shape (64, 16, 1) which means that sequences are reduced by half and the number of features is decreased by 80%.

Viewing the plots in the powerpoint presentation (slide 23), it seems that the model reliably predicts the earning rate at a similar accuracy and loss to the previous models.
However, when comparing epoch durations, the LSTM network of method `E3.1` takes half the time of what the model of `M5` takes, which goes to show that dimensionality reduction allows for comparably similar accuracy while having a more performant and efficient network in regards to computation due to a reduction of dimensionality.

```python
from torch.utils.tensorboard import SummaryWriter
import matplotlib.pyplot as plt
import time

# declare variables globally which will be accessed by training methods (among others)
net = None

hidden_size = 360
num_epochs = 20 # lower number of epochs due to time scarcity
batch_size = 64
num_layers = 3
num_directions = 1
learning_rate = 0.001

#loader = None
#valid_loader = None
#criterion = None
#optimizer = None
# setup tensorboard
tensorboard_path = None
#writer = None
# unique name for logging and plotting:
#model_name = None

features = ['Open', 'High', 'Low', 'Close', 'Volume']

def setup_training_AE_LSTM():
    global net, hidden_size, loader, criterion, optimizer, writer, valid_loader, model_name

    model_name = f'AE_layer2_LSTM_{int(time.time())}'
    print(f'Model name: {model_name}')

    dataset_train = StockDatasetTrain(features)
    loader = DataLoader(dataset=dataset_train, batch_size=batch_size, drop_last=True)
    dataset_valid = StockDatasetValidation(features)
    valid_loader = DataLoader(dataset=dataset_valid, batch_size=batch_size, drop_last=True)

    # Autenocder encodes down to 1 feature which will be input for the LSTM
    input_size = 1

    # init the LSTM model from the paper:
    # the autoencoder will be initialised in the start_training() function
    net = Net(input_size, hidden_size, num_layers, num_directions, batch_size)

    # loss function is (categorical) cross entropy
    # pytorch version already applies softmax in this loss function, which is why softmax is not implemented in the model
    criterion = nn.CrossEntropyLoss()
    # other available parameters for RMSprop (but not mentioned by the paper):
    # alpha=0.99, eps=1e-08, weight_decay=0, momentum=0, centered=False
    optimizer = optim.RMSprop(net.parameters(), lr=learning_rate)

    # init tensorboard
    # argument passed is a string defining the output directory for this specific run (which should be unique)
    tensorboard_path = f'{path_to_processed_data}/runs/train_{int(time.time())}_AE_LSTM/'
    tensorboard_path += f'Epochs{num_epochs}_Batch_sz{batch_size}_Hsz{hidden_size}_learning_rate{learning_rate}_AE_sigmoid_layers2'
    print(f'Tensorboard path: {tensorboard_path}')
    writer = SummaryWriter(tensorboard_path)
    
    start = time.time()
    accuracy, loss_values = start_training()
    end = time.time()
    print(f"{round(float(end - start),2)} seconds with hidden_sz : {hidden_size}")

    # Plot the loss after training
    plt.style.use('dark_background')
    plt.title(f'Hidden_sz: {hidden_size}; Epochs: {num_epochs}; Time: {round(float(end - start),2)}s')
    plt.plot(range(len(loss_values)), loss_values)
    plt.show()
    
    
setup_training_AE_LSTM()
```

# Conclusion
 
`M1` to `M5`:
Accuracy of a model is highly dependent on the input data (which stock, index, country etc.).
Having too few neurons / hidden layers can lead to loss of information while having too many increases complexity and brings up the vanishing gradient problem.

`E1` and `E2`:
Artificially extracting more features (by convolving the data) to train on may worsen accuracy and lead to more complex calculations and longer epoch durations as a result of receiving more data.

`E3`:
Determining the few features which optimally describe the variance of the given data yields efficient models of high accuracies.

With the results of this notebook it can be summed up that the given research papers results could have been improved by utilising different strategies to denoise the input data (such as implementing autoencoders), reduce its dimensionality and thereby allow the LSTM network to efficiently predict on the given data.

Other possible methods of improving the given model:
As proven in this notebook, dimensionality reduction shows best improvements. When combined with the principal component analysis, the features which most optimally desribe the variance can be selected to then pass through an encoder or a different strategie for further denoising of the data.
This way, computation is less complex due to the reduced size of the input data. This would then allow the inclusion of different types of data such as social media or news data related to given stock(s), which then can and should undergo a similar encoding method as mentioned above.

## Training for Method E1 - Increases Sequences
In this cell, the LSTM neural network defined in the paper is extended by a convolutional neural network.
This is the first extension `E1` of the research papers work.
The features which are passed to the network are:
`Open`, `High`, `Low`, `Close` and `Volume`
Data is normalized.
All hyperparameters which are used stayed the same as in previous cells.

The convolutional neural network part consists of 3 convolutional 1D layers where each layer is followed by a Maxpool 1D and a rectified linear function as activation.
The assumption behind this network is, that the more data a network processes, the better it is able to perform both on training and out of sample data. To have more data passed through the network, the CNN layers are supposed to extract more features and sequences from the actual input data by convolving the matrices and applying pooling. This way, the LSTM layers will receive data of size (64, 128, 5) instead of (64, 30, 5). This version of the network only increases sequences.

When looking at the plots in the powerpoint presentation (slide 20), one can observe that the model is able to generalize well. Validation loss falls quickly and remains stable throughout training.
However, having more sequences passed through to the network means that computations become more complex. The network is not able to improve accuracy more quickly than method M5. This can be explained by the greater amount of information which has to be processed. Not only does the network have to do accurate predictions based on the input data - it also has to extract the most relevant information and decide which data is of less importance at the same. This is why loss is higher and the epoch duration quadrupled (just as sequences quadrupled as well from 30 to 128).
In conclusion, using CNNs to expand input data even further does not improve performance. By extracting more data, computation takes longer which means that getting same accuracy takes more time when compared to `M5`.

## Training for Method E2 - Increases Sequences and Features
In this cell, the LSTM neural network defined in the paper is extended by a convolutional neural network.
This is the second extension `E2` of the research papers work.
The features which are passed to the network are:
`Open`, `High`, `Low`, `Close` and `Volume`
Data is normalized.
All hyperparameters which are used stayed the same as in previous cells.
Only now, compared to method `E1`: flag `incr_num_features` of the CNN network is set to true which means that features will increase as well.

The architecture of the network remains the same. By toggling the `incr_num_features` flag the output will change in a way where features are increased as well.
The assumption behind this network is, that increasing the number of features by convolving the input data, the network is able to perform at a higher level, similar to when number of features increased from method `M1` to `M5` where accuracy also increased as a result. The LSTM layers will now receive data of size (64, 128, 7) instead of (64, 30, 5). This version of the network increases both sequences and features.

Similar to `E1`, when looking at the plots in the powerpoint presentation (slide 21), one can observe that the model is able to generalize as well. Validation loss falls quickly and remains stable throughout training.
But again, having more sequences and features passed through to the network means that computations become more complex. 

All in all it shows that using methods to expand the size of the input data only seems to lead to greater complexity and longer computation time while performance experiences a slight decrease.
This is why the next extension aims to do the opposite with CNNs. Instead of expanding data further, extracting only the relevant information by applying dimensionality reduction and denoising may yield better performance for the network.
